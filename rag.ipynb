```python
import os

def read_text_files(directory='./data'):
    combined_text = ""
    
    if not os.path.exists(directory):
        raise FileNotFoundError(f"The directory '{directory}' does not exist.")
    
    for filename in os.listdir(directory):
        file_path = os.path.join(directory, filename)
        
        if os.path.isfile(file_path) and filename.endswith('.txt'):
            try:
                with open(file_path, 'r', encoding='utf-8') as file:
                    content = file.read()
                    combined_text += content.replace('\n', ' ')
            except Exception as e:
                print(f"Error reading {filename}: {str(e)}")
    
    return combined_text
```

```python
import pandas as pd

from langchain import PromptTemplate
from langchain.chains.question_answering import load_qa_chain
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
model = ChatGoogleGenerativeAI(model="gemini-1.5-flash", google_api_key=os.environ["GOOGLE_API_KEY"], temperature=0.2, convert_system_message_to_human=True)
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=10000, chunk_overlap=1000)
texts = text_splitter.split_text(context)
```

```python
embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001", google_api_key=os.environ["GOOGLE_API_KEY"])
```

```python
vector_index = Chroma.from_texts(texts, embeddings).as_retriever(search_kwargs={"k":5})
```

```python
with open('vectordb', 'w') as vdb:
    vdb.write(str(vector_index))
    vdb.close()
```

```python
qa_chain = RetrievalQA.from_chain_type(
    model, 
    retriever=vector_index,
    return_source_documents=True
)
```

```python
question = "paise kaise banaye"

result = qa_chain({'query':question})
```
